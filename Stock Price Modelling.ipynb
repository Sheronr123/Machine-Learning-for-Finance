```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('tesla.xlsx')
```

```python
df['Absolute Returns'] = df['Adj Close'].diff()
```

```python
df = df.set_index('Date')
df
```

```python
df[['Absolute Returns']].plot()
```

```python
df['Returns'] = df['Adj Close'].pct_change(1)
```

```python
df['Returns'].plot()
```

```python
df['Returns'].plot(kind='hist', bins=10)
plt.grid('on')
```

```python
df['Returns'].plot(kind='kde', bw_method=0.1)
df['Returns'].plot(kind='kde', bw_method=0.2)
df['Returns'].plot(kind='kde', bw_method=0.4)
```
